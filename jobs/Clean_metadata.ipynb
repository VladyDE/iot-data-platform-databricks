#### Script to clean all but one file in the metadata directory
The file will be the one that was used on the MERGE INTO script that's used to upsert the metadata

```python
import os

# We first obtain the name of the previous task in the job
# 'taskkey' has to be exactly the name of the previous task in the job
try:
    file_to_keep = dbutils.jobs.taskValues.get(
        taskKey="Update_Metadata", 
        key="latest_file_name", 
        default=""
    )
except Exception as e:
    print(f"No se pudo recuperar el Task Value: {e}")
    file_to_keep = ""

# Volume path
volume_path = "/Volumes/workspace/projects/iot_devices/iot_metadata/"

def cleanup_metadata_folder(path, keep_file):
    if not keep_file:
        raise ValueError("ERROR: El nombre del archivo a mantener llegó vacío. Abortando para evitar borrado masivo.")

    print(f"--- Iniciando Limpieza ---")
    print(f"Archivo a preservar: {keep_file}")

    files = os.listdir(path)
    deleted_count = 0
    
    for file_name in files:
        if file_name != keep_file:
            file_full_path = os.path.join(path, file_name)
            if os.path.isfile(file_full_path):
                os.remove(file_full_path)
                print(f"Eliminado: {file_name}")
                deleted_count += 1
    
    print(f"--- Proceso finalizado. Se eliminaron {deleted_count} archivos ---")

cleanup_metadata_folder(volume_path, file_to_keep)
```
